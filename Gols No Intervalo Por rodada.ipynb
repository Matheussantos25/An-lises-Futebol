```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo fornecido
file_path_atualizado = 'C:\\Users\\msdof\\OneDrive\\Imagens\\Ligas\\BASE DE DADOS GRANDES\\futpython.xlsx'
dados = pd.read_excel(file_path_atualizado)

# Filtrar dados específicos para a liga "Brazil Serie A" e a temporada "2022"
dados = dados[(dados['League'] == 'Brazil Serie A') & (dados['Season'].astype(str).str.contains('2022'))]

# Ordenar dados por data e rodada
dados = dados.sort_values(by=['Date', 'Rodada'])

# Agrupar por rodada e somar os gols no intervalo (TotalGoals_HT)
total_goals_ht_por_rodada = dados.groupby('Rodada')['TotalGoals_HT'].sum()

# Plotar
plt.figure(figsize=(10, 6))
total_goals_ht_por_rodada.plot(kind='bar')
plt.title('Total de Gols no Intervalo por Rodada - Brazil Serie A 2022')
plt.xlabel('Rodada')
plt.ylabel('Total de Gols no Intervalo')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo fornecido
file_path_atualizado = 'C:\\Users\\msdof\\OneDrive\\Imagens\\Ligas\\BASE DE DADOS GRANDES\\futpython.xlsx'
dados = pd.read_excel(file_path_atualizado)

# Filtrar dados específicos para a liga "Brazil Serie A" e a temporada "2022"
dados = dados[(dados['League'] == 'Brazil Serie A') & (dados['Season'].astype(str).str.contains('2022'))]

# Ordenar dados por data e rodada
dados = dados.sort_values(by=['Date', 'Rodada'])

# Filtrar os dados para TotalGoals_HT > 0
dados_com_gols_ht = dados[dados['TotalGoals_HT'] > 0]

# Contar o número de jogos com TotalGoals_HT > 0 por rodada
contagem_gols_ht_por_rodada = dados_com_gols_ht.groupby('Rodada').size()

# Criar um DataFrame para assegurar que todas as rodadas estão representadas
todas_rodadas = pd.DataFrame({'Rodada': range(1, dados['Rodada'].max() + 1)})
contagem_gols_ht_por_rodada = todas_rodadas.merge(contagem_gols_ht_por_rodada.reset_index(name='Contagem'), on='Rodada', how='left').fillna(0)

# Calcular a média do número de jogos com TotalGoals_HT > 0 por rodada
media_jogos_com_gols_ht = contagem_gols_ht_por_rodada['Contagem'].mean()

# Plotar
plt.figure(figsize=(10, 6))
plt.bar(contagem_gols_ht_por_rodada['Rodada'], contagem_gols_ht_por_rodada['Contagem'])
plt.axhline(y=media_jogos_com_gols_ht, color='r', linestyle='--', label=f'Média: {media_jogos_com_gols_ht:.2f}')
plt.title('Número de Jogos com Gols no Intervalo (> 0) por Rodada - Brazil Serie A 2022')
plt.xlabel('Rodada')
plt.ylabel('Número de Jogos com Gols no Intervalo')
plt.xticks(todas_rodadas['Rodada'])
plt.legend()
plt.show()

print(f'Média do número de jogos com TotalGoals_HT > 0 por rodada: {media_jogos_com_gols_ht:.2f}')
```
