```python
import pandas as pd
dados = pd.read_csv("C:\\Users\\msdof\\OneDrive\\Imagens\\Ligas\\BRA.csv")
dados_2023 = dados[dados['Season'] == 2023]
dados_2023
```

```python
# Copiar o data frame
dados_2023 = dados_2023.copy()
```

```python
dados_2023['Rodada'] = (dados_2023.index - dados_2023.index[0]) // 10 + 1
dados_2023
```

```python
dados_2023['Placar'] = dados_2023['HG'].astype(str) + '-' + dados_2023['AG'].astype(str)
dados_2023
```

HipÃ³tese de que determinado repetir um placar em casa na rodada seguinte Ã© pouco provÃ¡vel

```python
# Agrupando os dados por time da casa e rodada, depois shiftando os placares para comparar com a prÃ³xima rodada
dados_2023['Placar_Proxima_Rodada'] = dados_2023.groupby('Home')['Placar'].shift(-1)

# Calculando os casos em que o placar se repete
dados_2023['Placar_Repetido'] = dados_2023['Placar'] == dados_2023['Placar_Proxima_Rodada']

# Probabilidade de um time repetir o mesmo placar na rodada seguinte
probabilidade_repetir_placar = dados_2023['Placar_Repetido'].mean() * 100
probabilidade_repetir_placar
```

Testando Jogos desta HipÃ³tese

```python
# Encontrando os jogos seguintes onde o placar se repetiu
jogos_placar_repetido_com_proximo = dados_2023[dados_2023['Placar_Repetido']].copy()

# Adicionando informaÃ§Ãµes do jogo seguinte
jogos_placar_repetido_com_proximo['Data_Proximo_Jogo'] = jogos_placar_repetido_com_proximo.groupby('Home')['Date'].shift(-1)
jogos_placar_repetido_com_proximo['Adversario_Proximo_Jogo'] = jogos_placar_repetido_com_proximo.groupby('Home')['Away'].shift(-1)
jogos_placar_repetido_com_proximo['Placar_Proximo_Jogo'] = jogos_placar_repetido_com_proximo.groupby('Home')['Placar'].shift(-1)

```

```python
# Selecionando apenas os jogos onde o placar do prÃ³ximo jogo Ã© o mesmo
jogos_iguais_com_proximo = jogos_placar_repetido_com_proximo[jogos_placar_repetido_com_proximo['Placar'] == jogos_placar_repetido_com_proximo['Placar_Proximo_Jogo']]

# Exibindo os jogos
jogos_iguais_com_proximo = jogos_iguais_com_proximo[['Season', 'Rodada', 'Date', 'Home', 'Away', 'Placar', 'Data_Proximo_Jogo', 'Adversario_Proximo_Jogo', 'Placar_Proximo_Jogo']]
jogos_iguais_com_proximo
```

```python
# Contando o nÃºmero de jogos onde o placar se repetiu no prÃ³ximo jogo
numero_jogos_placar_repetido = jogos_iguais_com_proximo.shape[0]
numero_jogos_placar_repetido
```

```python
# Total de jogos com jogos consecutivos registrados
total_jogos_consecutivos = jogos_placar_repetido_com_proximo.shape[0]

# Jogos onde o placar nÃ£o se repetiu
jogos_placar_nao_repetido = total_jogos_consecutivos - numero_jogos_placar_repetido
jogos_placar_nao_repetido
```

HipÃ³tese de que determinado repetir um placar em casa nas 2 rodadas seguintes Ã© bem baixo

```python
# Shiftando os placares para os prÃ³ximos dois jogos
dados_2023['Placar_Proximo_2'] = dados_2023.groupby('Home')['Placar'].shift(-2)

# Verificando se o placar se repete por trÃªs jogos consecutivos
dados_2023['Placar_Repetido_3x'] = (dados_2023['Placar'] == dados_2023['Placar_Proxima_Rodada']) & (dados_2023['Placar'] == dados_2023['Placar_Proximo_2'])

# Probabilidade de um time repetir o mesmo placar em trÃªs jogos consecutivos
probabilidade_repetir_placar_3x = dados_2023['Placar_Repetido_3x'].mean() * 100
probabilidade_repetir_placar_3x
```

```python
# Contando o nÃºmero de jogos onde o placar se repetiu em trÃªs jogos consecutivos
numero_jogos_placar_repetido_3x = dados_2023['Placar_Repetido_3x'].sum()
numero_jogos_placar_repetido_3x
```

```python
# Filtrando os jogos onde o placar se repetiu em trÃªs jogos consecutivos
jogos_repetidos_3x = dados_2023[dados_2023['Placar_Repetido_3x']]

# Preparando os detalhes para exibiÃ§Ã£o
jogos_repetidos_3x = jogos_repetidos_3x[['Season', 'Rodada', 'Date', 'Home', 'Away', 'Placar', 'Placar_Proxima_Rodada', 'Placar_Proximo_2']]
jogos_repetidos_3x.head()
```

```python
# Criando condiÃ§Ã£o para jogos que repetiram o placar duas vezes mas nÃ£o trÃªs vezes
dados_2023['Placar_Repetido_2x'] = (dados_2023['Placar'] == dados_2023['Placar_Proxima_Rodada']) & (dados_2023['Placar'] != dados_2023['Placar_Proximo_2'])

# Filtrando esses jogos especÃ­ficos
jogos_repetidos_2x_nao_3x = dados_2023[dados_2023['Placar_Repetido_2x']]

# Preparando os detalhes para exibiÃ§Ã£o
jogos_repetidos_2x_nao_3x = jogos_repetidos_2x_nao_3x[['Season', 'Rodada', 'Date', 'Home', 'Away', 'Placar', 'Placar_Proxima_Rodada', 'Placar_Proximo_2']]
jogos_repetidos_2x_nao_3x

#tem NAN pq como tÃ¡ pegando sÃ³ dos mandantes e tem poucos jogos de mandantes, Ã© normal que tenha este comportamento

```

Independente do Mando

```python
# Criando uma nova coluna que identifica o time, independente de ser mandante ou visitante
dados_2023['Time'] = dados_2023['Home']
away_data = dados_2023[['Season', 'Rodada', 'Away', 'Placar']].rename(columns={'Away': 'Time'})
combined_data = pd.concat([dados_2023[['Season', 'Rodada', 'Time', 'Placar']], away_data])

# Ordenando por time e rodada
combined_data.sort_values(by=['Time', 'Season', 'Rodada'], inplace=True)

# Calculando o placar da prÃ³xima rodada para cada time
combined_data['Placar_Proxima_Rodada'] = combined_data.groupby('Time')['Placar'].shift(-1)

# Verificando se o placar se repete
combined_data['Placar_Repetido'] = combined_data['Placar'] == combined_data['Placar_Proxima_Rodada']

# Calculando a probabilidade de um time repetir o mesmo placar na prÃ³xima rodada, independente do mando
probabilidade_repetir_placar_indep = combined_data['Placar_Repetido'].mean() * 100
probabilidade_repetir_placar_indep
```
